# 12. Perform hypothesis testing to find the significant variables.


```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load the DataFrame
# df = pd.read_csv('your_data.csv')  # Load your actual data if not already loaded

# Define categorical and numerical variables
categorical_variables = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
numerical_variables = ['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi']

# Chi-square test for categorical variables
significant_categorical = {}
for col in categorical_variables:
    contingency_table = pd.crosstab(df[col], df['stroke'])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    if p < 0.05:  # significance level of 0.05
        significant_categorical[col] = p

# Correlation analysis for numerical variables
significant_numerical = {}
for col in numerical_variables:
    correlation = df[col].corr(df['stroke'])
    if abs(correlation) > 0.1:  # consider correlation coefficient with absolute value greater than 0.1 as significant
        significant_numerical[col] = correlation

# Print significant variables
print("Significant categorical variables:")
print(significant_categorical)
print("\nSignificant numerical variables:")
print(significant_numerical)
```

# We'll start with Logistic Regression.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Initialize the logistic regression model
logistic_model = LogisticRegression(random_state=42)

# Train the model on the resampled data
logistic_model.fit(X_train_resampled, y_train_resampled)

# Make predictions on the test set
y_pred = logistic_model.predict(X_test)

# Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

The classification report and confusion matrix provide an evaluation of the performance of a classification model. Here's what each metric means:

1. Precision: Precision is the ratio of true positive predictions to the total number of positive predictions made by the model. In this context, it represents the accuracy of the model in predicting positive cases (in this case, instances of stroke). A high precision indicates that the model makes fewer false positive predictions.

2. Recall: Recall, also known as sensitivity, is the ratio of true positive predictions to the total number of actual positive cases in the dataset. It measures the ability of the model to correctly identify positive cases. A high recall indicates that the model captures a large proportion of positive cases.

3. F1-score: The F1-score is the harmonic mean of precision and recall. It provides a balance between precision and recall. A high F1-score indicates that the model performs well in terms of both precision and recall.

4. Support: Support is the number of actual occurrences of each class in the dataset.

5. Accuracy: Accuracy is the ratio of correctly predicted instances to the total number of instances in the dataset. It measures the overall performance of the model in terms of correct predictions.

6. Confusion matrix: The confusion matrix is a table that summarizes the performance of a classification model. It shows the number of true positive, true negative, false positive, and false negative predictions made by the model.

Based on the provided classification report and confusion matrix:
- The model achieves high precision for the negative class (0), indicating that it accurately predicts instances of no stroke.
- The recall for the positive class (1) is lower, indicating that the model misses some instances of stroke.
- The F1-score for the positive class is also lower, reflecting the trade-off between precision and recall.
- The confusion matrix provides a detailed breakdown of the model's predictions, showing the number of true positive, true negative, false positive, and false negative predictions.

Overall, the model shows decent performance in predicting instances of stroke, with high precision for the negative class but lower recall and F1-score for the positive class.

# hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Define hyperparameters to tune
param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga']
}

# Initialize logistic regression model
logistic_model = LogisticRegression(max_iter=1000)

# Perform grid search with cross-validation
grid_search = GridSearchCV(logistic_model, param_grid, cv=5, scoring='f1', verbose=1)
grid_search.fit(X_train_resampled, y_train_resampled)

# Print best hyperparameters
print("Best Hyperparameters:", grid_search.best_params_)

# Make predictions on the test set
y_pred = grid_search.predict(X_test)

# Evaluate the model
from sklearn.metrics import classification_report, confusion_matrix
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

These results show the performance of the logistic regression model after tuning. The model achieves an accuracy of 80% on the test set, with a precision of 97% for the majority class (no stroke) and 15% for the minority class (stroke). The recall for the stroke class is 55%, indicating that the model is able to correctly identify approximately 55% of the actual stroke cases.

To address this issue, you may need to employ techniques such as resampling (e.g., oversampling the minority class, undersampling the majority class) or using different evaluation metrics (e.g., F1-score, precision-recall curve) that are more suitable for imbalanced datasets. Additionally, you could consider using more advanced algorithms or ensemble methods that are more robust to class imbalance, or performing hyperparameter tuning to optimize the model's performance.

The classification report and confusion matrix indicate the performance of the AdaBoost model:

- **Precision**: Precision for class 0 (no stroke) is high (0.94), indicating that when the model predicts a stroke is not present, it is correct 94% of the time. However, the precision for class 1 (stroke) is low (0.20), indicating that when the model predicts a stroke, it is correct only 20% of the time.
  
- **Recall**: Recall for class 0 is high (1.00), indicating that the model correctly identifies almost all instances of class 0. However, the recall for class 1 is very low (0.02), indicating that the model misses almost all instances of class 1.
  
- **F1-score**: The F1-score is a balance between precision and recall. The F1-score for class 0 is high (0.97), but for class 1, it is very low (0.03), reflecting the imbalance in precision and recall for class 1.

- **Accuracy**: The overall accuracy of the model is high (0.94), but it is misleading due to the imbalance in class distribution. The model achieves high accuracy by correctly predicting the majority class (class 0), but it performs poorly on the minority class (class 1).

- **Confusion Matrix**: The confusion matrix shows that the model correctly predicts the majority class (class 0) most of the time, but it struggles to correctly predict the minority class (class 1), as evidenced by the high number of false negatives (61).

Overall, while the model performs well in predicting the majority class, it performs poorly in identifying instances of the minority class. This imbalance in performance suggests that the model may need further tuning or that additional strategies, such as addressing class imbalance or exploring different algorithms, may be necessary to improve its performance on the minority class.
